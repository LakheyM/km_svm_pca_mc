```python
#Monte Carlo
import random
#Making Coin class

class Coin(object):
    sides=('heads', 'tails')
    last_result = None
    
    def flip(self):
        self.last_result=result=random.choice(self.sides)
        return result
def create_coins(number):
    return[Coin() for _ in range(number)]
def flip_coins(coins):
    for coin in coins:
        coin.flip()

def count_heads(flipped_coins):
    return sum(coin.last_result == 'heads' for coin in flipped_coins)

def count_heads(flipped_coins):
    return sum(coin.last_result == 'tails' for coin in flipped_coins)

#def main():
heads_count=[]
def run():
    
    coins = create_coins(1000)
    for i in range (100):
        flip_coins(coins)
        #print (count_heads(coins))
        #return (count_heads(coins))
        heads_count.append(count_heads(coins))
    return heads_count
        
if __name__ == '__main__':
    main()

#heads_count = []
#heads_count = run()
#heads_count
run()
```

```python
#scattering to visualize data as the trials proceed
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(range(len(heads_count)), heads_count)
#heads_count
```

```python
#Using binomial distribtion to study fair coin
n, p = 1000, .5  # number of trials, probability of each trial
s = np.random.binomial(n, p, 100)
# result of flipping a coin 100 times, tested 1000 times.
s

plt.hist(s)
```

```python
mean_s =np.mean(s)
var_s =np.var(s)
std_s= np.std(s)
max_s = np.max(s)
min_s = np.min(s)
print("mean: {}, Variance: {}, std: {}, max: {}, min: {}".format(mean_s, var_s, std_s, max_s, min_s))
```

```python
from numpy import array
range_list=[]
for i in range(1000):
    local_list=[]
    n, p = 1000, .5  # number of trials, probability of each trial
    s = np.random.binomial(n, p, 1000)
    # result of flipping a coin 1000 times, tested 1000 times.
    local_list.append(np.min(s))
    local_list.append(np.max(s))
    range_list.append(local_list)

range_list
a=array(range_list)
min_list=a[:, 0]
min_count = np.min(min_list)
max_list=a[:, 1]
max_count = np.max(max_list)
print("Counts of heads in 1000 trial of 1000 flips- Max: {} and Min: {}".format(max_count, min_count))
```
